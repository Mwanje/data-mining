```python
#get image by url of first image
resp = urllib.urlopen("https://upload.wikimedia.org/wikipedia/commons/8/8c/Cristiano_Ronaldo_2018.jpg")
CR7 = np.asarray(bytearray(resp.read()), dtype="uint8")
CR7 = cv2.imdecode(CR7, cv2.IMREAD_COLOR)
```

```python
# visualize the two test images
fig, ax = plt.subplots(1, 2)
clear_output()
ax[0].imshow(cv2.cvtColor(CR7, cv2.COLOR_BGR2GRAY), cmap='gray')
ax[0].axis('off')
ax[0].set_title('CR7')
ax[1].imshow(cv2.cvtColor(Baboon, cv2.COLOR_BGR2GRAY), cmap='gray')
ax[1].axis('off')
ax[1].set_title('Baboon')
```

# Face detection

```python
# path to the file containing the features that the openCv pipeline will look for in the frame
cascadePath = "Images/haarcascade_frontalface_default.xml"
# initialize a model for detecting whether a certain portion of an image contains a face
faceCascade = cv2.CascadeClassifier(cascadePath)
```

```python
#Convert the frame to grayscale. 
gray_CR7= cv2.cvtColor(CR7, cv2.COLOR_BGR2RGB)
```

```python
# openCv algorithm to scan the image looking for haar features
faces = faceCascade.detectMultiScale(
        gray_CR7,
        scaleFactor=1.1,
        minNeighbors=6,
        minSize=(100, 100),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
# output of the classifier
faces
```

```python
# draw a rectangle
from matplotlib.patches import Rectangle

# visualize the detected face
fig, ax = plt.subplots()
ax.imshow(gray_CR7, cmap='gray')

# iterate over all the detected faces
for face in faces:
    # retrieve the coordinates of the position of the current face, and its size    
    (x_CR7, y_CR7, w_CR7, h_CR7) = face
    # draw a rectangle where the face is detected    
    ax.add_artist(Rectangle((x_CR7, y_CR7), w_CR7, h_CR7, fill=False, lw=3, color='blue'))

ax.axis('off')
```

```python
gray_baboon = cv2.cvtColor(Baboon, cv2.COLOR_BGR2RGB)

faces = faceCascade.detectMultiScale(
        gray_baboon,
        scaleFactor=1.1,
        minNeighbors=6,
        minSize=(100, 100),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
# output of the classifier
faces
```

```python
fig, ax = plt.subplots()
ax.imshow(gray_baboon, cmap='gray')

# iterate over all the detected faces
for face in faces:
    # retrieve the coordinates of the position of the current face, and its size
    # SOLUTION:
    (x_baboon, y_baboon, w_baboon, h_baboon) = face
    
    # draw a rectangle where the face is detected
    # SOLUTION:
    ax.add_artist(Rectangle((x_baboon, y_baboon), w_baboon, h_baboon, fill=False, lw=3, color='yellow'))

ax.axis('off')
```
